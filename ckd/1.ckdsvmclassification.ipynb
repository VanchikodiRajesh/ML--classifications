```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
print (dataset)
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first="True")
```

```python
dataset
```

```python
input=dataset.drop(columns=['classification_yes'])
```

```python
output=dataset['classification_yes']
```

```python
input
```

```python
output
```

```python
from  sklearn.model_selection  import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input,output,test_size=0.2,random_state=2)
```

```python
from sklearn.svm import SVC
classi=SVC(kernel='rbf',random_state=0)
classi=classi.fit(x_train,y_train)
```

```python
y_pred=classi.predict(x_test)
```

```python

from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
print (cm)
```

```python
from sklearn.metrics import classification_report
```

```python
clf_report=classification_report(y_test,y_pred)
```

```python
print(clf_report)
```
